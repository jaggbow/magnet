# MPNN

```python
checkpoint_path = "/network/scratch/o/oussama.boussif/pdeone/logs/experiments/mpnn/runs/2022-04-20/04-50-47/checkpoints/epoch_186.ckpt"
```

```python
# Model specific args
model = FACTORY["mpnn"]
seed_everything(42)

trainer = pl.Trainer(gpus=1)
model = model.load_from_checkpoint(checkpoint_path).to(device).eval()
```

```python
testset = HDF5DatasetGraph( 
                path=test_path,
                mode='valid',
                nt=128,
                nx=256,
                in_timesteps=16,
                radius=2)
```

```python
batch = testset[nidx]
u = batch['u'].float().unsqueeze(0).to(device)
x = batch['x'].float() .unsqueeze(0).to(device)
B, N, _ = x.shape
t = batch['t'].float().unsqueeze(0).to(device)
```

```python
node_features, edge_index, edge_features = model._build_graph(u[:,:,:model.time_slice], x, t[:,:model.time_slice])
target = u[:,:,model.time_slice:]
T_out = target.shape[-1]
u_hat = []
for i in range(T_out//model.time_slice):
    y_hat = model(node_features, edge_index, edge_features)
    u_hat.append(y_hat)
    node_features = torch.cat([y_hat, x.reshape(B*N, -1), t[:, (i+2)*model.time_slice-1:(i+2)*model.time_slice].repeat(N, 1)], dim=-1)
    # inp = u_future[:,t*self.time_slice:(t+1)*self.time_slice]
    # inp = y_hat

u_hat = torch.cat(u_hat, dim=-1)
u_hat = u_hat.reshape(B, N, -1)

loss = model.criterion(u_hat, target)
mae_loss = model.mae_criterion(u_hat, target)
```

```python
u_mpnn = u_hat.squeeze().permute(1,0)
```

```python
print(mae_loss.item())
```

# NomeshGNN

```python
checkpoint_path = "/network/scratch/o/oussama.boussif/pdeone/logs/experiments/nomeshgnn/runs/2022-04-19/14-43-51/checkpoints/epoch_039.ckpt"
```

```python
# Model specific args
model = FACTORY["nomeshgnn"]
seed_everything(42)

trainer = pl.Trainer(gpus=1)
model = model.load_from_checkpoint(checkpoint_path).to(device).eval()
```

```python
testset = HDF5DatasetImplicit( 
                path=test_path,
                mode='valid',
                nt=128,
                nx=256,
                samples=32)
```

```python
batch = testset[nidx]
t = torch.from_numpy(batch['t']).float().unsqueeze(0).to(device)
u = batch['lr_frames'].float().unsqueeze(0).to(device) # B, T, 1, L
B, T, _, L = u.shape
u_values = batch['hr_points'].float().unsqueeze(0).to(device)
coords = batch['coords'].float().unsqueeze(0).to(device)
cells = batch['cells'].float().unsqueeze(0).to(device)
```

```python
u_values_future = u_values[:,model.time_slice:] # B, T_future, N, 1
T_future = u_values_future.shape[1]

u_values_hat = []
inp = u[:,:model.time_slice]
hr_last = u_values[:,model.time_slice-1]

for i in range(T_future//model.time_slice):
    y_hat, _ = model(inp, coords, cells, t[:,i*model.time_slice:(i+2)*model.time_slice], hr_last)
    u_values_hat.append(y_hat)

    inp = y_hat.permute(0,1,3,2)
    inp = F.interpolate(inp.reshape(-1,inp.shape[-2], inp.shape[-1]), size=L, mode='linear', align_corners=False).reshape(B, -1, inp.shape[-2], L)
    hr_last = y_hat[:,-1]

u_values_hat = torch.cat(u_values_hat, dim=1)
loss = model.criterion(u_values_hat, u_values_future)
mae_loss = model.mae_criterion(u_values_hat, u_values_future)
```

```python
u_nomeshgnn = u_values_hat.squeeze()
u_values_future = u_values_future.squeeze()
```

```python
print(mae_loss.item())
```

# FNO

```python
checkpoint_path = "/network/scratch/o/oussama.boussif/pdeone/logs/experiments/fno_1d/runs/2022-04-16/13-15-50/checkpoints/epoch_249.ckpt"
```

```python
# Model specific args
model = FACTORY["fno_1d"]
seed_everything(42)

trainer = pl.Trainer(gpus=1)
model = model.load_from_checkpoint(checkpoint_path).to(device).eval()
```

```python
testset = HDF5Dataset( 
                path=test_path,
                mode='valid',
                nt=128,
                nx=256)
```

```python
batch = testset[nidx]
u, dx, dt = batch
u = u.unsqueeze(0).to(device)
dx = dx.unsqueeze(0).to(device)
dt = dt.unsqueeze(0).to(device)
```

```python
u_history = u[:,:model.time_history] # B, T_history, L
u_future = u[:,model.time_history:] # B, T_future, L
T_future = u_future.shape[1]

u_hat = []
inp = u_history
for t in range((T_future//model.time_future)):
    y_hat = model(inp.permute(0,2,1), dx, dt).permute(0,2,1)
    u_hat.append(y_hat)
    inp = y_hat

u_hat = torch.cat(u_hat, dim=1)
u_fno = u_hat.squeeze()

loss = model.criterion(u_hat, u_future)
mae_loss = model.mae_criterion(u_hat, u_future)
```

```python
print(mae_loss.item())
```
