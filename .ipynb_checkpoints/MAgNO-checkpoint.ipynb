```python
import torch
from torch import nn
import numpy as np
import h5py
from torch_geometric.nn import knn_interpolate
import matplotlib.pyplot as plt

from xitorch.interpolate import Interp1D
```

```python
data = h5py.File('/network/scratch/o/oussama.boussif/pdeone/data/CE_train_E1_graph_50.h5', 'r')
```

```python
x_old = torch.from_numpy(data['train']['x'][0]).unsqueeze(1).float()
x_new = torch.linspace(0,16,100).unsqueeze(1).float()
u_old = torch.from_numpy(data['train']['pde_250-50'][10]).permute(1,0).float()

batch_old = torch.zeros(50).long()
batch_new = torch.zeros(100).long()
```

```python
u_knn = knn_interpolate(u_old, x_old, x_new, batch_x=batch_old, batch_y=batch_new, k=10, num_workers=1)
```

```python
plt.plot(x_old[:,0], u_old[:,-1], label='old')
plt.plot(x_new[:,0], u_knn[:,-1], label='knn')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,7))
ax[0].imshow(u_old.permute(1,0), cmap='RdBu_r', aspect="auto")
ax[1].imshow(u_knn.permute(1,0), cmap='RdBu_r', aspect="auto")
plt.show()
```
