# NomeshGraph

```python
checkpoint_path = "/network/scratch/o/oussama.boussif/pdeone/logs/experiments/nomeshgraph/runs/2022-05-14/16-46-02/checkpoints/epoch_249.ckpt"
```

```python
# Model specific args
model = FACTORY["nomeshgraph"]
seed_everything(42)

trainer = pl.Trainer(gpus=1)
model = model.load_from_checkpoint(checkpoint_path).to(device)
mse_criterion = nn.MSELoss(reduction='sum')
```

```python
nomeshgraph_results_mae = []
nomeshgraph_results_acc = []
nomeshgraph_outputs = []
ground_truth = []
for nx in [40,50,100,200]:
    print(f"nx={nx}")
    test_path = f'/home/mila/o/oussama.boussif/scratch/pdeone/data/CE_valid_E1_{nx}.h5'

    test_dataset = HDF5DatasetImplicitGNN( 
                    path=test_path,
                    mode='valid',
                    sampling='uniform',
                    nt=250,
                    nx=nx,
                    samples=12)
    dataloader = DataLoader(test_dataset, batch_size=32, num_workers=0, shuffle=False, pin_memory=True)
    model.eval()
    mae_loss = 0
    acc_mse = 0
    with torch.no_grad():
        for batch in tqdm(dataloader, leave=False):
            t = batch['t'].float().to(device)
            u = batch['lr_frames'].float().to(device) # B, T, 1, L
            u_values = batch['hr_points'].float().to(device)
            coords = batch['coords_hr'].float().to(device)
            lr_coords = batch['coords_lr'].float().to(device)
            B = t.shape[0]

            u_values_future = u_values[:,model.time_slice:] # B, T_future, N, 1
            T_future = u_values_future.shape[1]

            u_values_hat = []
            inp = u[:,:model.time_slice]
            hr_last = u_values[:,model.time_slice-1]

            for i in range(T_future//model.time_slice):
                out_hr, out_lr, _ = model.forward(
                    inp, 
                    lr_coords, 
                    coords, 
                    t[:,i*model.time_slice:(i+2)*model.time_slice], 
                    hr_last)
                y_hat = torch.cat([out_hr, out_lr], dim=2)
                u_values_hat.append(y_hat)

                inp = out_lr.permute(0,1,3,2)
                hr_last = out_hr[:,-1]

            u_values_hat = torch.cat(u_values_hat, dim=1)
            target = torch.cat([u_values_future, u[:,model.time_slice:].permute(0,1,3,2)], dim=2)

            mae_loss += model.mae_criterion(u_values_hat, target)*B/(len(test_dataset))
            acc_mse += mse_criterion(u_values_hat, target)/(nx*len(test_dataset))
    
    nomeshgraph_results_mae.append(mae_loss.item())
    nomeshgraph_results_acc.append(acc_mse.item())
    
    nomeshgraph_outputs.append(u_values_hat[0].squeeze().cpu())
    ground_truth.append(target[0].squeeze().cpu())
```

```python
nomeshgnn_results = []
nomeshgnn_outputs = []
for nx in [40,50,100,200]:
    test_path = f'/home/mila/o/oussama.boussif/scratch/pdeone/data/CE_valid_E1_{nx}.h5'

    test_dataset = HDF5DatasetImplicit( 
                    path=test_path,
                    mode='valid',
                    sampling='uniform',
                    nt=250,
                    nx=nx,
                    samples=12)
    dataloader = DataLoader(test_dataset, batch_size=32, num_workers=0, shuffle=False, pin_memory=True)
    model.eval()
    mae_loss = 0
    with torch.no_grad():
        for batch in tqdm(dataloader, leave=False):
            t = batch['t'].float().to(device)
            u = batch['lr_frames'].float().to(device) # B, T, 1, L
            B, T, _, L = u.shape
            u_values = batch['hr_points'].float().to(device)
            coords = batch['coords'].float().to(device)
            cells = batch['cells'].float().to(device)
            B = t.shape[0]

            u_values_future = u_values[:,model.time_slice:] # B, T_future, N, 1
            T_future = u_values_future.shape[1]

            u_values_hat = []
            inp = u[:,:model.time_slice]
            hr_last = u_values[:,model.time_slice-1]

            for i in range(T_future//model.time_slice):
                y_hat, _, _ = model.forward(inp, coords, cells, t[:,i*model.time_slice:(i+2)*model.time_slice], hr_last)

                u_values_hat.append(y_hat)

                inp = y_hat.permute(0,1,3,2)
                inp = F.interpolate(inp.reshape(-1,inp.shape[-2], inp.shape[-1]), size=L, mode='linear', align_corners=False).reshape(B, -1, inp.shape[-2], L)
                hr_last = y_hat[:,-1]

            u_values_hat = torch.cat(u_values_hat, dim=1)
            loss = model.criterion(u_values_hat, u_values_future)
            mae_loss += model.mae_criterion(u_values_hat, u_values_future)*B/len(test_dataset)
    nomeshgnn_results.append(mae_loss.item())
    nomeshgnn_outputs.append(u_values_hat[0].squeeze().cpu())
```

```python
checkpoint_path = '/network/scratch/o/oussama.boussif/pdeone/logs/experiments/mpnn/runs/2022-05-04/18-03-22/checkpoints/epoch_248.ckpt'
# Model specific args
model = FACTORY["mpnn"]
seed_everything(42)

trainer = pl.Trainer(gpus=1)
model = model.load_from_checkpoint(checkpoint_path).to(device)
```

```python
mpnn_outputs = []
mpnn_results = []
for nx in [40,50,100,200,400,600,800,1000,2000]:
    test_path = f'/home/mila/o/oussama.boussif/scratch/pdeone/data/CE_valid_E1_{nx}.h5'

    test_dataset = HDF5DatasetGraph( 
                    path=test_path,
                    mode='valid',
                    nt=250,
                    nx=nx,
                    in_timesteps=25,
                    radius=1)
    dataloader = DataLoader(test_dataset, batch_size=32, num_workers=0, shuffle=False, pin_memory=True)
    model.eval()
    mae_loss = 0
    with torch.no_grad():
        for batch in tqdm(dataloader):
            u = batch['u'].float().to(device)
            x = batch['x'].float().to(device)
            B, N, _ = x.shape
            t = batch['t'].float().to(device)

            node_features, edge_index, edge_features = model._build_graph(u[:,:,:model.time_slice], x, t[:,:model.time_slice])
            target = u[:,:,model.time_slice:]
            T_out = target.shape[-1]
            u_hat = []
            for i in range(T_out//model.time_slice):
                y_hat = model.forward(node_features, edge_index, edge_features)
                u_hat.append(y_hat)
                node_features = torch.cat([y_hat, x.reshape(B*N, -1), t[:, (i+2)*model.time_slice-1:(i+2)*model.time_slice].repeat(N, 1)], dim=-1)
                # inp = u_future[:,t*self.time_slice:(t+1)*self.time_slice]
                # inp = y_hat

            u_hat = torch.cat(u_hat, dim=-1)
            u_hat = u_hat.reshape(B, N, -1)

            loss = model.criterion(u_hat, target)
            mae_loss += model.mae_criterion(u_hat, target)*B/len(test_dataset)
    mpnn_results.append(mae_loss.item())
    mpnn_outputs.append(u_hat[0].squeeze().cpu())
```

```python
fig, ax = plt.subplots(1,7, figsize=(20,5))
for i in range(7):
    ax[i].imshow(mpnn_outputs[i].permute(1,0), cmap='RdBu_r', aspect="auto")
plt.show()
```

# FNO

```python
checkpoint_path = '/network/scratch/o/oussama.boussif/pdeone/logs/experiments/fno_1d/runs/2022-05-04/13-32-38/checkpoints/epoch_244.ckpt'
# Model specific args
model = FACTORY["fno_1d"]
seed_everything(42)

trainer = pl.Trainer(gpus=1)
model = model.load_from_checkpoint(checkpoint_path).to(device)
```

```python
fno_results = []
fno_outputs = []
ground_truth = []

for nx in [40,50,100,200,400,600,800,1000,2000]:
    test_path = f'/home/mila/o/oussama.boussif/scratch/pdeone/data/CE_valid_E1_{nx}.h5'

    test_dataset = HDF5Dataset( 
                    path=test_path,
                    mode='valid',
                    nt=250,
                    nx=nx)
    dataloader = DataLoader(test_dataset, batch_size=32, num_workers=0, shuffle=False, pin_memory=True)
    model.eval()
    mae_loss = 0
    with torch.no_grad():
        for batch in tqdm(dataloader):
            u, dx, dt = batch
            u = u.float().to(device)
            dx = dx.float().to(device)
            dt = dt.float().to(device)

            L = u.shape[-1]

            u_history = u[:,:model.time_history] # B, T_history, L
            u_future = u[:,model.time_history:] # B, T_future, L
            T_future = u_future.shape[1]

            u_hat = []
            inp = u_history
            for t in range(T_future//model.time_future):
                y_hat = model.forward(inp.permute(0,2,1), dx, dt).permute(0,2,1)
                u_hat.append(y_hat)
                inp = y_hat

            u_hat = torch.cat(u_hat, dim=1)

            loss = model.criterion(u_hat, u_future)
            mae_loss += model.mae_criterion(u_hat, u_future)*B/len(test_dataset)
    fno_results.append(mae_loss.item())
    fno_outputs.append(u_hat[0].squeeze().cpu())
    ground_truth.append(u_future[0].squeeze().cpu())
```

```python
fig, ax = plt.subplots(2,7, figsize=(20,10))
for i in range(7):
    ax[0,i].imshow(fno_outputs[i], cmap='RdBu_r', aspect="auto")
    ax[1,i].imshow(ground_truth[i], cmap='RdBu_r', aspect="auto")
plt.show()
```
