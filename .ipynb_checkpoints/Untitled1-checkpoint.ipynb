```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
results = [[0.1815, 0.1989, 0.20346, 0.16279, 0.1953],
    [0.16428,0.152831,0.12807,0.13202,0.117],
    [0.1035,0.10224,0.10588,0.09654,0.1425],
    [0.0899,0.09459,0.090907,0.08853,0.09255],
    [0.1242,0.108778,0.09342,0.0811,0.116337]]
results = np.array(results)
```

```python
plt.rcParams.update({'font.size': 22})

plt.figure(figsize=(10,7))
plt.errorbar([2,4,8,16,25], results.mean(-1), yerr=results.std(-1), marker='o')
plt.grid()
plt.ylabel('MAE')
plt.xlabel('Number of training spatial queries')
plt.savefig('spatial_queries.svg')
plt.show()
```
