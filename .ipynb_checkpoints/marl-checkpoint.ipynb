```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

```python
ippo = [
    pd.read_csv('/home/mila/o/oussama.boussif/NPS_GWT_DVNC/MARL/PettingZoo/pettingzoo_simple_spread_algo_IPPO_agents_3_seed_0/2022_05_14_22_16/training_rewards.csv'),
    pd.read_csv('/home/mila/o/oussama.boussif/NPS_GWT_DVNC/MARL/PettingZoo/pettingzoo_simple_spread_algo_IPPO_agents_3_seed_42/2022_05_14_21_37/training_rewards.csv'),
    pd.read_csv('/home/mila/o/oussama.boussif/NPS_GWT_DVNC/MARL/PettingZoo/pettingzoo_simple_spread_algo_IPPO_agents_3_seed_2022/2022_05_14_21_43/training_rewards.csv')
]
cppo = [
    pd.read_csv('/home/mila/o/oussama.boussif/NPS_GWT_DVNC/MARL/PettingZoo/pettingzoo_simple_spread_algo_CPPO_agents_3_seed_0/2022_05_14_22_25/training_rewards.csv'),
    pd.read_csv('/home/mila/o/oussama.boussif/NPS_GWT_DVNC/MARL/PettingZoo/pettingzoo_simple_spread_algo_CPPO_agents_3_seed_42/2022_05_14_22_19/training_rewards.csv'),
    pd.read_csv('/home/mila/o/oussama.boussif/NPS_GWT_DVNC/MARL/PettingZoo/pettingzoo_simple_spread_algo_CPPO_agents_3_seed_2022/2022_05_14_22_19/training_rewards.csv')
]
end = 318403
```

```python
ippo_aggregate = 0
train_steps = ippo[0]['train_steps'][:end]
for i in range(3):
    ippo_aggregate += ippo[i]['train_rewards'][:end]/3
    
cppo_aggregate = 0
train_steps = cppo[0]['train_steps'][:end]
for i in range(3):
    cppo_aggregate += cppo[i]['train_rewards'][:end]/3
```

```python
N = 1000
ippo_smooth = np.convolve(ippo_aggregate.values, np.ones((N,))/N, mode='valid') 
cppo_smooth = np.convolve(cppo_aggregate.values, np.ones((N,))/N, mode='valid') 
```

```python
plt.plot(train_steps[:-N+1], ippo_smooth, label='IPPO')
plt.plot(train_steps[:-N+1], cppo_smooth, label='CPPO')
plt.legend()
plt.grid()
plt.show()
```
